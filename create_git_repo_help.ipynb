### В данной инструкции мы:
1. Создадим репозиторий на своей машине
2. Добавим **email** и **name** в глобальный конфиг git'а (делается только один раз, все дальнейшие репозитории будут использовать эти конфигурационные данные)
3. Закоммитим файл hello_world.txt в данном локальном репозитории
4. Создадим удаленный репозиторий на GitHub (нужно предварительно зарегистрироваться)
5. Добавим remote локальному репозиторию (свяжем удаленный и локальный репозитории)
6. Запушим hello_world.txt
7. Рекомендация

### Шаг 1


* Заходим в терминал (на windows это **cmd**, на linux и macos есть соответствующее стандартное приложение)  
* Создаем папочку с названием **dev** (например, но лучше хранить весь свой код в определенной папке, удобно назвать ее dev - development)  
```bash
mkdir dev  # Создаем папку
cd dev  # Заходим в эту папку
```
* Создаем еще одну папку внутри папки **dev**, которую потом превратим в гит репозиторий
```bash
mkdir test_git_repo
cd test_git_repo
```
* Инициализуем **git**
```bash
git init
```


### Шаг 2


* Добавим имя и почту в конфигурационный файл Git
```bash
git config --global user.name "Vasiliy Pupkin"  # Сюда запишите свое имя и фамилию через пробел (лучше использовать реальные имя и фамилию, а не никнейм)
git config --global user.email "vasya@thebest.com"  # Ваша почта, по которой потом зарегистрируетесь на GitHub
```


### Шаг 3


* Создаем текстовый файл hello_world.txt (можете в принципе любой другой файл создать, в том числе питоновский скрипт)
```bash
touch hello_world.txt  # Команда touch создает пустой файл с названием hello_world.txt в данной директории
echo "Hello world" > hello_world.txt  # Добавляем текст "Hello world" в файл hello_world.txt
```
* Добавляем изменения репозитория на сцену (на белый лист)
```bash
git add --all  # Закидываем все модифицированные и новые файлы на сцену
```
* Зафиксируем изменения (сворачиваем лист и клеим на него надпись, чтобы идентифицировать в дальнейшем)
```bash
git commit -m "Commit message"  # Commit message - любой текст, который объясняет какие изменения в данной версии вы внесли
```

### Шаг 5


* Возвращаемся в терминал
* Добавляем remote локальному репозиторию
```bash
git remote add origin https://github.com/leokul01/python_networks.git  # Не забудьте заменить ссылку на свою
```


### Шаг 6


* Пушим локальные изменения в удаленный репозиторий
```bash
git push --set-upstream origin master  # Флаг set-upstream нужен только при первом пуше (он связывает ветку master с удаленной веткой мастер)
```
